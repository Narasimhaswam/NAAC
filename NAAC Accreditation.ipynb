# NAAC
India has one of the largest and diverse education systems in the world. Privatization, widespread expansion, increased autonomy and introduction of Programmes in new and emerging areas have improved access to higher education. At the same time, it has also led to widespread concern on the quality and relevance of the higher education. To address these concerns, the National Policy on Education (NPE, 1986) and the Programme of Action (PoA, 1992) spelt out strategic plans for the policies, advocated the establishment of an independent National accreditation agency. Consequently, the National Assessment and Accreditation Council (NAAC) was established in 1994 as an autonomous institution of the University Grants Commission (UGC) with its Head Quarter in Bengaluru. The mandate of NAAC as reflected in its vision statement is in making quality assurance an integral part of the functioning of Higher Education Institutions (HEIs).

## How does it work?
The assessment and accreditation process of NAAC involves the following steps:

Preparation of Self-Study Report (SSR): Institutions applying for accreditation prepare an SSR, which provides comprehensive information about their infrastructure, academic programs, faculty, research, governance, and other relevant aspects.

Peer Team Visit: NAAC appoints a team of experts, known as the Peer Team, to visit the institution and validate the information provided in the SSR. The Peer Team comprises academic and administrative experts from various fields.

Evaluation: During the visit, the Peer Team assesses the institution based on the predefined assessment criteria set by NAAC. They interact with faculty, staff, students, and other stakeholders to gather information and evaluate the institution's overall quality.

Report and Accreditation: After the visit, the Peer Team prepares a report highlighting the strengths and weaknesses of the institution. Based on the report, NAAC decides the accreditation status of the institution.

## Criteria of Assessment:
NAAC assesses institutions based on seven criteria, each with several sub-criteria. These criteria provide a comprehensive framework for evaluating the overall quality of the institution. The seven criteria are as follows:

1. Curricular Aspects: Focuses on the design, delivery, and innovation in academic programs.
2. Teaching-Learning and Evaluation: Emphasizes effective teaching, learning methodologies, and evaluation systems.
3. Research, Innovations, and Extension: Assesses the research output, innovations, and extension activities of the institution.
4. Infrastructure and Learning Resources: Evaluates the adequacy and quality of infrastructure, facilities, and learning resources.
5. Student Support and Progression: Considers the support services provided to students and their overall progression in academic and co-curricular activities.
6. Governance, Leadership, and Management: Examines the governance structure, leadership effectiveness, and management practices.
7. Institutional Values and Best Practices: Looks at the promotion of core institutional values and the adoption of best practices.

## Criteria & Weightages :
NAAC has identified a set of seven criteria to serve as the basis of its assessment procedures. NAAC has categorized the Higher Educational Institutions into three major types (University, Autonomous College, and Affiliated/Constituent College) and assigned different weightages to these criteria under different key aspects based on the functioning and organizational focus of the three types of HEIs.

Key Indicators and Weightages

The criterion-wise differential weightages for the three types of HEIs are:

Curricular Aspects 150 (U) 150 (Au) 100 (Aff UG) 100 (Aff PG)

Teaching-learning & Evaluation 200 (U) 300 (Au) 350 (Aff UG) 350 (Aff PG)

Research, Innovations & Extension 250 (U) 150 (Au) 110 (Aff UG) 120 (Aff PG)

Infrastructure & Learning Resources 100 (U) 100 (Au) 100 (Aff UG) 100 (Aff PG)

Student Support & Progression 100 (U) 100 (Au) 140 (Aff UG) 130 (Aff PG)

Governance, Leadership & Management 100 (U) 100 (Au) 100 (Aff UG) 100 (Aff PG)

Institutional Values & Best Practices 100 (U) 100 (Au) 100 (Aff UG) 100 (Aff PG)

Key Indicators Under each Criterion a few Key Indicators are identified. These Key Indicators (KIs) are further delineated as Metrics which actually elicit responses from the HEIs.

## Grading System and How is it decided:
NAAC assigns a Cumulative Grade Point Average (CGPA) to the institution, which represents the overall performance. The grading system and how it is decided are as follows:

CGPA Calculation: The CGPA is calculated by dividing the total weighted score obtained by the institution by the maximum weighted score.

Letter Grade: Based on the CGPA, a letter grade is assigned, ranging from A++ to C, with A++ being the highest and C being the lowest.

The accreditation status and grade awarded by NAAC are valid for a specific period, usually ranging from one to five years,
